# **Usando Pandas para Web Scrapping** 

## **URL da página usada neste exemplo se chama: "2018-19 NBA Player Stats: Per Game"**

Atualmente existem varias formas de se fazer web scrapping com Python por exemplo: Modulo urllib e Beautifulsoup. Mas a uma função mais simples que os dois exemplos citados acima. Pandas possui uma função chamada pd.read_html para ler as paginas na web.
Com essa função conseguimos extrair os dados de uma página direto para o Jupyter Notebook.

Neste caso usarei uma tabela de dados de pontuação de jogadores da NBA durante duas temporadas. E uma analise simples dos dados pois o objetivo principal é mostrar como a função pd.read_html pode ser pratico e fácil de ultizar no seu codigo.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ano = '2019'
url_link = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'

url = url_link.format(ano)
url
```

```python
anos = [2018,2019]
url_link = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'

for ano in anos:
  url = url_link.format(ano)
  print(url)
```

```python
url = 'https://www.basketball-reference.com/leagues/NBA_2019_per_game.html'
```

## **Read HTML**

### Usando a função de Pandas "pd.read_html()" para ler a página em HTLM

```python
df = pd.read_html(url, header = 0)
df
```

```python
len(df)
```

```python
df[0]
```

```python
df2019 = df[0]
```

## Data Cleaning

```python
df2019[df2019.Age == 'Age']
```

```python
len(df2019[df2019.Age == 'Age'])
```

```python
df = df2019.drop(df2019[df2019.Age == 'Age'].index)
```

```python
df.shape
```

## Exemplo de visualização simples de dados

#### Histogram

```python
plt.figure(figsize=(16,9))
sns.distplot(df.PTS)

plt.title("Pontos", fontsize=17)
plt.xlabel("Pontos Feitos por Jogador", fontsize=15)
plt.ylabel("Numero de Jogadores", fontsize=15)
plt.show()
```
